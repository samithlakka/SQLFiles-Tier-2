```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
database = "sqlite_db_pythonsqlite.db"
connection = sql.connect(database)
```

```python
#Q10: Produce a list of facilities with a total revenue less than 1000.
#The output of facility name and total revenue, sorted by revenue. Remember
#that there's a different cost for guests and members!
query = """SELECT bookid, Facilities.name, SUM(CASE WHEN memid > 0 THEN membercost ELSE guestcost END) AS Cost FROM Bookings LEFT JOIN Facilities USING (facid) LEFT JOIN Members USING (memid) GROUP BY name HAVING Cost < 1000 ORDER BY name"""
df = pd.read_sql_query(query, connection)
df
```

```python
#Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */
query = """SELECT (member.surname || ' ' || member.firstname) AS member_name,
       (recommended.surname || ' ' || recommended.firstname) AS recommended_name
FROM Members AS member
LEFT JOIN Members AS recommended ON member.recommendedby = recommended.memid
WHERE member.recommendedby > 0
ORDER BY member.surname, member.firstname;"""
df = pd.read_sql_query(query, connection)
df
```

```python
#Q12: Find the facilities with their usage by member, but not guests */
query = """
SELECT (surname || ' ' || firstname) AS member, 
       Facilities.name, 
       COUNT(Facilities.name) AS Count 
FROM Facilities 
JOIN Bookings USING (facid) 
LEFT JOIN Members USING (memid) 
WHERE memid > 0 
GROUP BY member, Facilities.name
"""
df = pd.read_sql_query(query, connection)
df
```

```python
#Q13: Find the facilities usage by month, but not guests */
#query = """SELECT Facilities.name, Bookings.starttime, EXTRACT(YEAR FROM Bookings.starttime) AS year FROM Facilities JOIN Bookings USING (facid)"""
query = """SELECT Facilities.name, strftime('%Y - %m', Bookings.starttime) AS month, COUNT(memid) AS usage FROM Facilities JOIN Bookings USING (facid) WHERE memid > 0 GROUP BY name, month"""
df = pd.read_sql_query(query, connection)
df
```
